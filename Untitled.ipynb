### Formatando

```python
def values_color(value):
    color = 'red' if value < 3 else 'blue'
    return 'color: %s'%color #color : 'red' ou color : 'blue'

dados[['gold', 'silver', 'bronze']].head(10).style.applymap(values_color)
#Apenas nos 10 elementos serie head() se fosse em todos
```

```python
#ou poderia usar assim
dados.head(10).style.applymap(values_color, subset=['gold', 'silver', 'bronze'])
```

```python
#Destacando maior valor em cada coluna especificada
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

dados.head(10).style.apply(highlight_max, subset=['gold', 'silver', 'bronze'])
```

```python
#applymap em todos os elemento(elemento a elemento) / apply em toda a coluna
#combinando as duas formatações

dados[['gold', 'silver', 'bronze']].head(10).style.\
applymap(values_color).\
apply(highlight_max)
```

```python
#somente colunas especificadas iguais
dados[dados['gold']==dados['silver']]
```

```python
dados.filter(['country', 'gold', 'silver', 'bronze'])
```

```python
#criando input 
gold_median = dados['gold'].median()
dados.query('gold > @gold_median')
```

```python
#Pegando somente o que esta sendo especificado
dados[dados['country'].isin(['Germany', 'Brazil'])]
```

```python
#Somente quem não é
dados[~dados['country'].isin(['Germany', 'Brazil'])]
```

```python
#Com base no tipo
numericos = ['int32', 'int64']
dados.select_dtypes(include= numericos)
```

```python
#o contrario
dados.select_dtypes(exclude= numericos)
```
